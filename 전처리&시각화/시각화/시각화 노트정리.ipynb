# 그래프를 그리는 기준
 - 통계분석에서도 동일한 기준으로 분석한다.
1. 자료의 형태(변수의 종류)
  - 범주형(문자로 되어있는 경우)자료와 연속형(숫자로 되어있는 경우) 자료
  - 질적변수(범주형)와 양적변수(연속형)
  - 질적변수 : 변수와 같이 비 수치적이고, 특정 카테고리에 포함되는 경우
    - 명목형 : 특정 범주에 포함되지만, 순위를 갖지 않는 경우( 성별, 혈액형 등)
    - 순위형 : 특정 범주에 포함되지만, 순위를 갖는 경우 (성적, 등급 등)
  - 양적 변수 : 변수의 값이 수치적이고, 연속적인 경우(키, 나이)
    - 이산형 : 셀 수 있는 경우 (정수인 경우)
    - 연속형 : 셀 수 없는 경우 (실수인 경우)
    - → 알고리즘을 돌리려면 다 양적 변수로 바꿔야함

2. 자료의 차원
  - 변수의 개수는 차원이 된다.
  - 차원이 높으면, 시각화를 해둘 수 없다.
    - 최대 2차원의 화면에서 3차원까지만 나타낼 수 있다.
  - 1차원
    - 일변수(일변량) : 변수 하나에 대한 시각화
    - 변수의 분포 등을 확인 
  - 2차원
    - 이변수(이변량) : 변수와 변수의 관계를 확인하고 싶은 경우
      - 연속형 변수와 연속형 변수
      - 연속형 변수와 이산형 변수
      - 이산형 변수와 이산형 변수

#### **상관계수** 
**: 연관성을 표현하는 수치**
 * 피어슨 상관게수 ( -1 ~1까지의 값을 가짐 )
  - 상관계수가 **0인 경우** 
    - **전혀 연관성 없음**
  - 상관계수가 **1인 경우** 
    - **확실한 연관성 있음**
  - 상관계수가 **-1인 경우**
    - **확실한 반대 관계**
  - **상관계수 > 0 일 경우**
    - **1에 가까울수록 연관성이 강해짐**
    - x와 y가 비례
  - **상관계수 < 0**
    - **-1에 가까울수록 연관성이 더 강해짐**
    - x와 y가 반비례

전체 콜레스테롤 수치와 트리글리세라이드 수치 차이에는 크게 관계가 없어보임

rugplot은 자료의 위치 정도를 확인할 수 있다. 밀도 정도는 확인 가능
자료가 매우 많은 경우에는 `히스토그램`을 확인할 수 없는 경우가 있는데, 그런 경우 `rugplot`을 이용하면 빠르게 확인 가능

```python
# rugplot - 실수 분포 플롯
# 자료의 분포를 묘사하기 위한 것 /  데이터 위치를 x축 위에 작은 선분(rug)으로 나타내어 실제 데이터의 위치를 보여줌
sns.rugplot(data=health, x='총콜레스테롤',height= 1)
```

'IDL콜레스테롤' 수치를 구해보자
  - 파생변수라고 한다.
    - 주어진 자료에서 target변수를 가장 잘 설명할 수 있는 새로운 변수를 만들어내는 것 → `데이터마이닝`
    - 분석에서 가장 중요한 부분
  - IDL콜레스테롤 = 총콜레스테롤 - (HDL콜레스테롤 + LDL콜레스테롤)

```python
tmp['총콜레스테롤']-( tmp['HDL콜레스테롤'] + tmp['LDL콜레스테롤'])
```

```python
tmp['IDL콜레스테롤'] = tmp['총콜레스테롤']-( tmp['HDL콜레스테롤'] + tmp['LDL콜레스테롤'])
tmp
```

```python
sns.jointplot(data=tmp, x='IDL콜레스테롤', y='트리글리세라이드')
```

#### boxplot
- 탐색적 자료 분석(EDA)에서 주로 많이 사용
- 4분위수를 이용한 시각화
  - 이상치(outlier)를 확인하는 용도로 사용
  - 자료를 4개의 구간으로 나눈 값
    - 1분위수: 자료의 25%에 해당하는 값
    - 2분위수: 자료의 50%에 해당하는 값(중앙값과 동일)
    - 3분위수: 자료의 75%에 해당하는 값

![](https://i.imgur.com/tssRkgA.jpg)

- IQR(Inter Quantile Range)
  - IQR = Q3 - Q1
  - 최대제한선 = Q3 + (IQR * 1.5)

  ![](https://i.imgur.com/CnFsXL9.png)
