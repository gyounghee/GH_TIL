## json
- 'javaScript Object Notation'
  - 자바스크립트에서 객체를 표현하는 방법 처음에 제안
  - 요새는 언어와 상관없이 데이터를 전달하는 표준적인 방법으로 많이 쓰임
    - 이러한 방법을 '객체 직렬화'라고 함
  - 네트워크를 통해서 객체를 전달할 수 없기 때문에 객체를 전달하는 다양한 방법들이 존재
    - json은 문자 전송 방식 → json자체는 일반 문자열이라고 보면 됨
  
  - 파이썬의 'dict' 타입과 호환성이 좋음
    - 변환이 쉽다(dict→json, json→dict)
    - dict → json : json.dumps()
    - json → dict : json.loads()
    - json도 `키:값`쌍으로 표현되는 문자열

### 데이터 파싱
- 일자별 지역별 → 확진자, 사망자 수

````python
for row in result['response']['body']['items']['item'] :
  print('날짜 :', row['createDt'])
  print('지역 :', row['gubun'])
  print('확진자 :', row['deathCnt'])
  print('사망자 :', row['defCnt'])
````

날짜별 지역별로 테이블을 생성
```
{ '2020-01-01' : ['서울', 100], ['강원' : 20] ...}  // 첫 번째 방법
{ '2020-01-01' : {'서울' : 100}, {'강원' : 20}... }  // 두 번째 방법
```

````python
# 날짜를 키로하고 지역별 확진자를 값으로 하는 딕셔너리
confirmed = { }
for row in result['response']['body']['items']['item'] :
  key = row['createDt'].split(' ')[0]

  # 해당 날짜의 가장 마지막 지역의 확진자 수가 값으로 들어가는 문제가 생김
  # 해당 날짜의 키가 존재하지 않으면 딕셔너리를 추가 , 존재하면 값만 추가
  if key in confirmed.keys() :
    confirmed[key][row['gubun']] = row['defCnt']         
  else : 
    confirmed[key] = { }
    confirmed[key][row['gubun']] = row['defCnt']   
````

### 일자별 시간대별로 데이터를 요청해야 받아올 수 있음

````python
!pip install xmltodict
import requests
import xmltodict
````

````python
key = '456e594e6b67686c37377974455758'
spot = 'D-42'  # 테헤란로(선릉역) 기준으로 확인할거임
````

````python
# 날짜를 키로하고 vol을 값으로 하는 딕셔너리
traffic = {}
for day in range(1,32):  # 1일부터 31일까지
  for hour in range(0,24):  # 00시부터 23시까지
    # :02를 통해 두자리로 자릿수를 맞춰줌
    url = f'http://openapi.seoul.go.kr:8088/{key}/xml/VolInfo/1/8/{spot}/202107{day:02}/{hour:02}/'
    response = requests.get( url )
    result = xmltodict.parse( response.text )
    status = result['VolInfo']['RESULT']['CODE']
    if status == 'INFO-000':   # 정상 처리가 되었으면
      vol = 0
      for row in result['VolInfo']['row']:
        vol += int( row['vol'] )   # 전체 교통량 합
      # print('{}-{}-{:02}'.format( row['ymd'][4:6], row['ymd'][6:], int(row['hh'])+1 ) )   # 날짜 →  월 일 시
      # print(vol)
      dt = '{}-{}-{:02}'.format( row['ymd'][4:6], row['ymd'][6:], int(row['hh']) + 1)
      traffic[dt] = vol
````

````python
traffic
````

````python
import pandas as pd
````

````python
pd.Series( traffic )
````

````python
traffic_vol = pd.DataFrame.from_dict( traffic, orient = 'index', columns=['vol'])
traffic_vol
````

````python
# 컬럼으로 뺴려면 인덱스를 reset시켜주면됨
traffic_vol.reset_index(inplace = True)
traffic_vol.rename( columns={'index' :'date'}, inplace = True )
````

### 유입/유출 구분

````python
key = '456e594e6b67686c37377974455758'
spot = 'D-42'  # 테헤란로(선릉역) 기준으로 확인할거임
````

````python
# 날짜를 키로하고 vol을 값으로 하는 딕셔너리
from collections import defaultdict
traffic = defaultdict(dict)

for day in range(1,32):  # 1일부터 31일까지
  for hour in range(0,24):  # 00시부터 23시까지
    # :02를 통해 두자리로 자릿수를 맞춰줌
    url = f'http://openapi.seoul.go.kr:8088/{key}/xml/VolInfo/1/8/{spot}/202107{day:02}/{hour:02}/'
    response = requests.get( url )
    result = xmltodict.parse( response.text )
    status = result['VolInfo']['RESULT']['CODE']
    if status == 'INFO-000':   # 정상 처리가 되었으면
      in_vol = 0
      out_vol = 0
      for row in result['VolInfo']['row']:
        if row['io_type'] == '1' : in_vol += int( row['vol'] )   # 전체 유입 교통량 합
        if row['io_type'] == '2' : out_vol += int( row['vol'] )   # 전체 유출 교통량 합
      # print('{}-{}-{:02}'.format( row['ymd'][4:6], row['ymd'][6:], int(row['hh'])+1 ) )   # 날짜 →  월 일 시
      dt = '{}-{}-{:02}'.format( row['ymd'][4:6], row['ymd'][6:], int(row['hh']) + 1)
      traffic[dt]['in'] = in_vol
      traffic[dt]['out'] = out_vol
````

````python
traffic
````

````python
import pandas as pd
````

````python
traffic_inout_vol = pd.DataFrame( traffic ).T
traffic_inout_vol.reset_index(inplace=True)
traffic_inout_vol
````
