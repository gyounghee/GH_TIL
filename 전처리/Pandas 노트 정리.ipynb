### 다중조건
 - and, or, not
  - Pandas는 &(앰퍼샌드), |(파이프라인), ~(틸드) 문자를 사용해서 표현
  - 괄호를 이용하여 우선순위를 명확하게 하는 것이 좋음

````python
# 서초구 지점의 최현경 담당자의 자료만 검색
rawData[ (rawData['지점']=='서초구') & (rawData['담당'] == '최현경')]
````

````python

rawData[(rawData['지점']=='서초구') & ((rawData['담당']=='최현경') | (rawData['담당']=='김현경'))]
````

### **문자열 처리**

````python
# isin은 리스트를 파라미터로 갖는다.
# 리스트 내의 값들 중에서 하나라도 존재하면 True, 그렇지 않으면 False
rawData[(rawData['지점']=='서초구') & (rawData['담당'].isin(['김현경','최현경']))]
````

````python
# 담당자들 중에 성이 '김'씨인 자료만 검색
# contains는 문자열 내에서 특정 문자열이 존재하면 True, 그렇지 않으면 False
rawData[ rawData['담당'].str.contains('김')]
````

````python
# startswith는 문자열 내에서 특정 문자열로 시작하는 경우 True, 그렇지 않으면 False
rawData[rawData['담당'].str.startswith('김')]
````

````python
# 담당자의 성이 김으로 시작하지 않는 자료 검색
rawData[ ~(rawData['담당'].str.startswith('김'))]
````

### 이상치 (outlier)
 - 특별히 크거나, 작은 값
  - 얼마나 크거나 작아야 이상치라고 할 수 있을까? → 분석가의 주관
 - 이상치를 찾고자 하는 경우가 아니라면, 이상치는 제거하고 분석을 진행한다.

````python
rawData['몸무게'].describe()
````

````python
# 특정 값을 넘어가는 자료를 이상치로 판단해보면
rawData[rawData['몸무게'] > 76.15]
````

````python
# 4분위 수를 이용한 방법
display(rawData['몸무게'].quantile(0.25) )
display(rawData['몸무게'].quantile(0.50) )
display(rawData['몸무게'].quantile(0.75) )
display(rawData['몸무게'].quantile(0.99) )
````

````python
# 이상치를 상위 99%를 넘아가는값
# 하위 0.1보다는 낮은 값
# 이상치를 제거하는 방법
low = rawData['몸무게'].quantile(0.01)
high = rawData['몸무게'].quantile(0.99)
rawData[ (rawData['몸무게'] > low) & (rawData['몸무게'] < high) ]
````

## apply
 - Pandas  Dataframe(데이터프레임, 2차원) 타입의 객체에서 사용 가능한 반복문 
 - 모든 행, 열에 대해서 동일한 함수를 반복적으로 적용

````python
# apply의 기본적인 사용 방법
def func(x):
  print("함수 호출")
  print(x)

  # apply를 통해서 호출되는 함수는 return을 꼭 작성
  return

# 기본적으로는 DataFrame의 각 columns들의 시리즈 형태로 전달
rawData.apply(func)
````

````python
def func(x):
  print()
  print("함수 호출")
  print(x)
  print('-----------------------------')

  return

rawData.apply(func, axis = 1)   # axis = 0 이면 컬럼,  1이면 행
````

````python
# 지점별로 지역을 새로 만들어보자
def func(x):
  if x['지점'] == '관악구' : return '강서'
  elif x['지점'] == '여의도' : return '강서'
  elif x['지점'] == '강남구' : return '강동'
  elif x['지점'] == '서초구' : return '강동'

# apply가 return하는 값 들도 하나의 Series가 된다.
# 이를 이용해서 새로운 컬럼을 하나 만드는 것도 가능
rawData.apply(func, axis=1)
````

## inner join
 - merge 함수를 통해서 `join`을 해볼 수 있다.
   - merge의 기본 동작은 `inner join`이다.
 - inner join은 기본적으로 자료의 크기가 훨씬 줄어들게 된다.
  - user_usage : 240개 / user_device : 272개
  - 이 두자료를 하나로 합쳤을 경우 최대 240이다.
  - 양쪽 자료 모두에 존재하지 않는 자료가 있다면, 최소는 0이 될 수도 있다.

````python
display(len(user_usage))
display(len(user_device))
````

````python
pd.merge(left=user_usage, right=user_device, on='use_id')
````

## left join
 - left 자료를 기준으로 합쳐준다.
 - 결측치가 발생할 수 있음
  - right에 없는 자료는 결측치로 채워진다.

````python
# how의 기본값은 inner 
pd.merge(left=user_usage, right=user_device, on='use_id', how = 'left')  
````

## full outer join

````python
pd.merge(left=user_usage, right=user_device, on='use_id', how = 'outer') 
````
